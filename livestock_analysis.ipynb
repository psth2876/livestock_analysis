```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
store_yak = pd.read_csv('yak-nak-chauri-population-in-nepal-by-district.csv')
store_wool = pd.read_csv('wool-production-in-nepal-by-district.csv')
store_rabbit = pd.read_csv('rabbit-population-in-nepal-by-district.csv')
store_production_egg = pd.read_csv('production-of-egg-in-nepal-by-district.csv')
store_production_cotton = pd.read_csv('production-of-cotton-in-nepal-by-district.csv')
store_meat = pd.read_csv('net-meat-production-in-nepal-by-district.csv')
store_milk = pd.read_csv('milk-animals-and-milk-production-in-nepal-by-district.csv')
store_horse = pd.read_csv('horseasses-population-in-nepal-by-district.csv')
```

```python
store_yak['DISTRICT']= store_yak['DISTRICT'].replace(['SANKHUWASHAVA'],
                            'SANKHUWASABHA').replace(['E. REGION'],
                            'E.REGION').replace(['C. REGION'],
                            'C.REGION').replace(['W. REGION'],
                            'W.REGION').replace(['MW. REGION'],
                            'MW.REGION').replace(['FW. REGION'],
                            'FW.REGION').replace(['RAMECHAP'],
                            'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_yak.head()
```

```python
store_wool['DISTRICT']= store_wool['DISTRICT'].replace(['SANKHUWASHAVA'],
                                  'SANKHUWASABHA').replace(['E. REGION'],
                                  'E.REGION').replace(['C. REGION'],
                                  'C.REGION') .replace(['W. REGION'],
                                  'W.REGION').replace(['MW. REGION'],
                                  'MW.REGION').replace(['FW. REGION'],
                                  'FW.REGION').replace(['RAMECHAP'],
                                  'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_wool
```

```python
store_rabbit['DISTRICT']= store_rabbit['DISTRICT'].replace(['SANKHUWASHAVA'],
                                      'SANKHUWASABHA').replace(['E. REGION'],
                                    'E.REGION').replace(['C. REGION'],
                                    'C.REGION').replace(['W. REGION'],
                                    'W.REGION').replace(['MW. REGION'],
                                     'MW.REGION').replace(['FW. REGION'],
                                    'FW.REGION').replace(['RAMECHAP'],
                                     'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_rabbit.head()
```

```python
store_production_egg['DISTRICT']= store_production_egg['DISTRICT'].replace(['SANKHUWASHAVA'],
                                                       'SANKHUWASABHA').replace(['E. REGION'],
                                                       'E.REGION').replace(['C. REGION'],
                                                       'C.REGION').replace(['W. REGION'],
                                                       'W.REGION').replace(['MW. REGION'],
                                                       'MW.REGION').replace(['FW. REGION'],
                                                       'FW.REGION').replace(['RAMECHAP'],
                                                       'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_production_egg.head()
```

```python
store_production_cotton['DISTRICT'] = store_production_cotton['DISTRICT'].str.upper()
store_production_cotton
```

```python
store_meat['DISTRICT']= store_meat['DISTRICT'].replace(['SANKHUWASHAVA'],
                                               'SANKHUWASABHA').replace(['E. REGION'],
                            'E.REGION').replace(['C. REGION'],
                            'C.REGION').replace(['W. REGION'],
                            'W.REGION').replace(['MW. REGION'],
                            'MW.REGION').replace(['FW. REGION'],
                            'FW.REGION').replace(['RAMECHAP'],
                            'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_meat
```

```python
store_milk['DISTRICT']= store_milk['DISTRICT'].replace(['SANKHUWASHAVA'],
                                             'SANKHUWASABHA').replace(['E. REGION'],
                            'E.REGION').replace(['C. REGION'],
                            'C.REGION').replace(['W. REGION'],
                            'W.REGION').replace(['MW. REGION'],
                            'MW.REGION').replace(['FW. REGION'],
                            'FW.REGION').replace(['RAMECHAP'],
                            'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_milk.head(10)
```

```python
store_horse['DISTRICT']= store_horse['DISTRICT'].replace(['SANKHUWASHAVA'],
                                     'SANKHUWASABHA').replace(['E. REGION'],
                                     'E.REGION').replace(['C. REGION'],
                                     'C.REGION').replace(['W. REGION'],
                                     'W.REGION').replace(['MW. REGION'],
                                     'MW.REGION').replace(['FW. REGION'],
                                     'FW.REGION').replace(['RAMECHAP'],
                                     'RAMECHHAP').replace(['TERATHUM'],'TERHATHUM')
store_horse.head()
```

```python
df_merge1=pd.merge(store_milk,store_horse,on='DISTRICT',how='outer')
df_merge1                                                                        
```

```python
df_merge2=pd.merge(store_meat,store_production_cotton,on='DISTRICT',how='outer')
df_merge2
```

```python
df_merge3=pd.merge(store_production_egg,store_rabbit,on='DISTRICT',how='outer')
df_merge3
```

```python
df_merge4=pd.merge(store_wool,store_yak,on='DISTRICT',how='outer')
df_merge4
```

```python
df_merge5=pd.merge(df_merge1,df_merge2,on='DISTRICT',how='outer')
df_merge5
```

```python
df_merge6=pd.merge(df_merge3,df_merge4,on='DISTRICT',how='outer')
df_merge6
```

```python
df_final=pd.merge(df_merge5,df_merge6,on='DISTRICT',how='outer')
df_final
```

```python
df_last=df_final.drop([96]) # drop removes the specified rows from dataframe. # drop Total row because the value in it is incorrect. 
```

```python
western_region= df_last[df_last['DISTRICT'].str.startswith(("BAG", "MUST", "MYA", "PAR", "GOR", "KAS", "LAM", "MANA", "SYAN", "TANA", "ARGA", "GUL", "KAPI", "NAWA", "PAL","RUPA"))]
western_region
```

```python
#Total horse population in Western Development Region in Nepal
horse_population = western_region[['DISTRICT','Horses/Asses']]

#Reshaping
horse_reshape= pd.melt(horse_population, id_vars="DISTRICT",value_name="Horse Population",var_name="Horse")
plt.figure(figsize=(20,5))

# barplots to visualize the data
sns.barplot(x="DISTRICT", y="Horse Population", data= horse_reshape, hue="Horse", palette="Set2")
plt.xlabel('Districts of Western Development Region')
```

## Net Meat Production according to Regions of Nepal

```python
net_meat= df_last[df_last['DISTRICT'].str.endswith(("REGION"))]
net_meat
```

```python
#Total meat production in different district in Nepal
meat_production = net_meat[['DISTRICT','BUFF','MUTTON','CHICKEN','CHEVON']]

#Reshaping
meat_reshape= pd.melt(meat_production, id_vars="DISTRICT",value_name="Meat Production",var_name="ANIMALS")
plt.figure(figsize=(20,5))
#plot the point plot
sns.pointplot(x="DISTRICT", y="Meat Production", data= meat_reshape, hue="ANIMALS")
plt.title('Bar Diagram to show animals meat production in different district of Nepal')
plt.xlabel('Development Regions of Nepal')
```

## Heatmap for Total production of items 

```python
#correlation between total production of different animals and items
corr_rel= df_last[['TOTAL MILK PRODUCED','TOTAL EGG','TOTAL MEAT','SHEEP WOOL PRODUCED']].corr()
plt.figure(figsize=(10,7))

sns.heatmap(corr_rel,annot =True,cmap="hot_r")
plt.title('Heatmap figure for Production of Different Items in Different District',fontsize=12)
plt.xlabel('Production Items',fontsize=10)
plt.ylabel('Production Items',fontsize=10)
```

```python
rabbit_analysis= df_final[df_final['DISTRICT'].str.startswith(("FW.","MW.","W.","E.","C."))]
rabbit_analysis.head()
```

```python
# Total rabbit numbers in different development region of Nepal.
rabbit_pop = rabbit_analysis[['DISTRICT','Rabbit']]

# Reshaping.
rabbit_reshape = pd.melt(rabbit_pop, id_vars = "DISTRICT", value_name = "Rabit Population", var_name = 'Rabbit')
plt.figure(figsize=(20,5))

# Line Plot.
sns.lineplot(x ="DISTRICT", y ="Rabit Population", data = rabbit_reshape, hue = 'Rabbit', palette="Set2")

# Adding Title.
plt.title('Line Plot Rabbit Population in Differents Regions of Nepal') 
plt.xlabel('All categorical Regions of Nepal')
```

```python
cotton_analysis= df_final[df_final['DISTRICT'].str.startswith(("DANG","BANKE","BARDIYA"))]
cotton_analysis
```

```python
# Total cotton production in different districts of Nepal.
cotton_production = cotton_analysis[['DISTRICT','AREA (Ha.)','PROD. (Mt.)','YIELD Kg/Ha']]

# Reshaping.
cotton_reshape = pd.melt(cotton_production, id_vars = "DISTRICT", value_name = "Cotton YIELD Production", var_name = 'Cotton Type')
plt.figure(figsize=(20,5))

# Line Plot.
sns.lineplot(x ="DISTRICT", y ="Cotton YIELD Production", data = cotton_reshape, hue = 'Cotton Type', palette="Set2")

# Adding Title.
plt.title('Line PLot Cotton Production in Different districts of Nepal') 
plt.xlabel('Districts of Nepal')
```
